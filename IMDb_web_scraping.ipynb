```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
genres = input("Enter the genres: ")

# construct the search URL with the genres parameter
url = f"https://www.imdb.com/search/title?genres={genres}"
```

```python
# Send request to the URL and get the response
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the movie containers
movies = soup.find_all('div', class_='lister-item mode-advanced')

# Create empty lists to store the data
titles = []
stars = []
summaries = []
release_dates = []
writers = []
directors = []

# Loop through the movies and extract the data
for movie in movies:
    # Title
    title = movie.h3.a.text
    titles.append(title)

    # Stars
    stars_text = ''
    stars_tags = movie.find_all('a', href=lambda href: href and '/name/' in href)
    for i, tag in enumerate(stars_tags):
        if i > 0:
            stars_text += ', '
        stars_text += tag.text
    stars.append(stars_text)

    # Summary
    summary = movie.find_all('div', class_='lister-item-content')[0].find_all('p')[1].text.strip()
    summaries.append(summary)

    # Release date
    release_date = movie.find_all('span', class_='lister-item-year text-muted unbold')[0].text.replace('(', '').replace(')', '').strip()
    release_dates.append(release_date)

    # Writer
    writer_tag = movie.find_all('p', class_='')[0].find_all('a')[0]
    writer = writer_tag.text
    writers.append(writer)

    # Director
    director_tag = movie.find_all('p', class_='')[0].find_all('a')[1]
    director = director_tag.text
    directors.append(director)

# Create a dataframe with the extracted data
df = pd.DataFrame({'Title': titles, 'Stars': stars, 'Summary': summaries, 'Release Date': release_dates, 
                   'Writer': writers, 'Director': directors})
```
